3.Define Machine Learning. What are the main components in Machine Learning?

Definition of Machine Learning (ML):

Machine Learning is a branch of Artificial Intelligence (AI) that focuses on building systems which can learn patterns from data and improve performance automatically without being explicitly programmed.

**Main Components of Machine Learning:**

Dataset – The collection of data used to train and test the model.

Features (Input variables) – The measurable properties or characteristics fed into the model.

Model/Algorithm – The mathematical or computational method that learns patterns from data (e.g., Decision Tree, Neural Network).

Training – The process of feeding data to the model so it can learn the relationships.

Evaluation – Measuring how well the model performs using metrics (accuracy, precision, recall, etc.).

Prediction – Using the trained model to make decisions or predictions on new, unseen data.

4.How does loss value help in determining whether the model is good or not?

Loss Value (or Cost Function):

The loss value tells us how far off the model’s predictions are from the actual/true values.

It’s a number calculated during training that the model tries to minimize.

**it helps determine model quality:**

Lower loss = better model fit

If the loss is very small, the model’s predictions are close to actual values → good model.

Higher loss = poor model fit

If the loss is large, predictions are far from the true values → bad model.

Trend during training matters

If the loss decreases steadily, the model is learning.

If the loss stays high or fluctuates, the model is not learning well.

7.What do you mean by training and testing a dataset?

Training a Dataset


Training means using a portion of the dataset to teach the model.

The model looks at the input data (features) and the correct answers (labels/targets), and tries to learn patterns/relationships.

Example: A model learns from house prices (size, location → price).

Testing a Dataset

Testing means using a separate portion of the dataset (not seen during training) to check how well the model has learned.

It helps us evaluate the model’s performance on new, unseen data.

Example: After training, we test the model with new house data to see if it predicts the price correctly.

10.How do we split data for model fitting (training and testing) in Python?
How do you approach a Machine Learning problem?

we approach machine learning problem by:

Problem → Data → Preprocessing → EDA → Model Selection → Training → Evaluation → Deployment → Monitoring

```python
from sklearn.model_selection import train_test_split

# Example dataset
X = [[1], [2], [3], [4], [5], [6]]   # Features
y = [10, 20, 30, 40, 50, 60]         # Labels

# Split data: 70% training, 30% testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("Training data:", X_train, y_train)
print("Testing data:", X_test, y_test)
```

15.What is causation? Explain difference between correlation and causation with an example.

Causation means one variable directly affects another (cause → effect).

Correlation means two variables move together but without proving cause.

Example:

Correlation: Ice cream sales ↑ and drowning cases ↑ in summer (both rise due to heat, not because ice cream causes drowning).

Causation: Smoking ↑ → Lung cancer risk ↑ (direct effect).

21.What is feature scaling? How does it help in Machine Learning?

Feature scaling is the process of bringing all features to a similar scale (range).

Why it helps:

Prevents features with large values from dominating.

Speeds up gradient descent convergence.

Improves performance of distance-based models (KNN, SVM).
